```python
import tensorflow as tf
import keras
from keras import *
from keras.layers import *
from keras.models import *
import keras.backend as K
import numpy as np
import time
import pandas as pd
import os
import matplotlib.pyplot as plt
```

```python
def create_model(num_layers,num_width):
#     model = Sequential()
    
    ## inputs
    t = Input(shape=(1,))
    x = Input(shape=(1,))
    y = Input(shape=(1,))
    
    X = concatenate([t, x, y], 1)
#     X = 2.0*(tf.concat([t,x,y], 1) - X_min)/(X_max - X_min) - 1.0
#     t_tmp = 2.0*(t - self.X_min[0])/(self.X_max[0] - self.X_min[0]) - 1
    
    ## eta_model
    first_layer = Dense(num_width,activation='relu')(t)
#     new_layer = Dense(num_width,activation='relu')(first_layer)    
    out_eta = Dense(1,activation='sigmoid',name='out_eta')(first_layer)
    
    ## uvp model
    first_layer_uvp = Dense(num_width,activation='relu')(X)
    new_layer_uvp = Dense(num_width,activation='relu')(first_layer_uvp)
    for l in range(int((num_layers-2)/2)-1):
        new_layer_uvp = BatchNormalization()(new_layer_uvp)
        new_layer_uvp = Dense(num_width,activation='relu')(new_layer_uvp)
        new_layer_uvp = Dense(num_width,activation='relu')(new_layer_uvp)
    
    out_uvp = Dense(3,activation='sigmoid',name='out_uvp')(new_layer_uvp)
    
    
    model = Model(inputs=[t,x,y] , outputs=[out_eta ,out_uvp])
    
    return model
```

```python
model = create_model(10,32)
model.summary()
```

```python
def fwd_gradients(U, x):
    layers_uvp = [3] + 10*[3*32] + [3]
    dummy_tf = tf.placeholder(tf.float32, shape=(None, layers_uvp[-1]))
    g = tf.gradients(U, x, grad_ys=dummy_tf)[0]
    return tf.gradients(g, dummy_tf)[0]
```

```python
def net_VIV(txy,y_pred,Re=1000):
    
#         X = 2.0*(tf.concat([t,x,y], 1) - self.X_min)/(self.X_max - self.X_min) - 1.0
#         uvp = model.outputs[0:3] #self.neural_net(X, self.weights_uvp, self.biases_uvp)

#         t_tmp = 2.0*(t - self.X_min[0])/(self.X_max[0] - self.X_min[0]) - 1
#         eta = model.outputs[3:4] #self.neural_net(t_tmp, self.weights_eta, self.biases_eta)
            
        uvp = y_pred[:3]
        u = uvp[0:1]
        v = uvp[1:2]
        p = uvp[2:3]
        eta = y_pred[3:4]
    
        txy = model.inputs
        t = txy[0:1]
        x = txy[1:2]
        y = txy[2:3]

        uvp_t = fwd_gradients(uvp, t)
        uvp_x = fwd_gradients(uvp, x)
        uvp_y = fwd_gradients(uvp, y)
        uvp_xx = fwd_gradients(uvp_x, x)
        uvp_yy = fwd_gradients(uvp_y, y)

        eta_t = tf.gradients(eta, t)[0]
        eta_tt = tf.gradients(eta_t, t)[0]

#         u = uvp[0:1]
#         v = uvp[1:2]
#         p = uvp[2:3]
#         eta = uvp[3:4]

        u_t = uvp_t[:,0:1]
        v_t = uvp_t[:,1:2]

        u_x = uvp_x[:,0:1]
        v_x = uvp_x[:,1:2]
        p_x = uvp_x[:,2:3]

        u_y = uvp_y[:,0:1]
        v_y = uvp_y[:,1:2]
        p_y = uvp_y[:,2:3]

        u_xx = uvp_xx[:,0:1]
        v_xx = uvp_xx[:,1:2]

        u_yy = uvp_yy[:,0:1]
        v_yy = uvp_yy[:,1:2]

        eq1 = u_t + (u*u_x + v*u_y) + p_x - (1.0/Re)*(u_xx + u_yy)
        eq2 = v_t + (u*v_x + v*v_y) + p_y - (1.0/Re)*(v_xx + v_yy) + eta_tt
        eq3 = u_x + v_y

#         return u, v, p, eta, eq1, eq2, eq3
        
#         def custom_loss(y_true,y_pred):
        loss = tf.reduce_mean(tf.square(eq1)) + tf.reduce_mean(tf.square(eq2)) + tf.reduce_mean(tf.square(eq3))
#             return loss
        
        return loss
```

```python
def custom_loss_wrapper(input_tensor):
#     input_tensor = concatenate([t, x, y], 1)
    def custom_loss(y_true, y_pred):
        navier_loss = net_VIV(input_tensor,y_pred,Re=1000)
        return tf.reduce_mean(tf.square(y_true - y_pred)) + navier_loss
    return custom_loss

model.compile(loss={'out_eta': 'mean_squared_error', 'out_uvp': custom_loss_wrapper(model.layers[3])}, optimizer='adam', metrics=['accuracy','mean_squared_error'])
```

```python
def load_data(filename = "final_translated_data.csv"):
    time_1 = time.time()
    data = pd.read_csv("final_translated_data.csv")
    print("[INFO] Time taken = "+str(time.time()-time_1)+" seconds.")
    return data
```

```python
def create_tensors(data):
    
    t_star = data['t'] # T x 1
    eta_star = data['eta'] # T x 1
    
    T = t_star.shape[0]
        
    X_star = data['x']
    Y_star = data['y']        
    U_star = data['u']
    V_star = data['v']
    P_star = data['p']

    t = t_star.to_numpy()
    eta = eta_star.to_numpy()
    
    # T = t_star.shape[0]
        
    x = X_star.to_numpy()
    y = Y_star.to_numpy()        
    u = U_star.to_numpy()
    v = V_star.to_numpy()
    p = P_star.to_numpy()
    
    t = t.reshape((t.shape[0],1))
    x = x.reshape((x.shape[0],1))
    y = y.reshape((y.shape[0],1))
    u = u.reshape((u.shape[0],1))
    v = v.reshape((v.shape[0],1))
    p = p.reshape((p.shape[0],1))
    eta = eta.reshape((eta.shape[0],1))
    
    return t,x,y,u,v,p,eta
```

```python
def plot_time_data(time_snap):
  time_data = data[data['t']==time_snap]
  x_snap = time_data["x"]
  y_snap = time_data["y"]
  plt.plot(x_snap, y_snap, 'bo',markersize=0.1,marker='o',scalex=3,scaley=3)
  plt.title("at time: "+str(time_snap)+" second")
  plt.show()
```

```python
data = load_data(filename="final_translated_data.csv")
t,x,y,u,v,p,eta = create_tensors(data)
t.shape,x.shape,y.shape,u.shape,v.shape,p.shape,eta.shape
```

```python
N_train = 3800000
idx = np.random.choice(t.shape[0], N_train, replace=False)
t_train = t[idx]
x_train = x[idx]
y_train = y[idx]
u_train = u[idx]
v_train = v[idx]
p_train = p[idx]
eta_train = eta[idx]
t_train.shape,x_train.shape,y_train.shape,u_train.shape,v_train.shape,p_train.shape,eta_train.shape
```

```python
for time in np.arange(start=1,step=0.5,stop=3.02):
  plot_time_data(time)
```

```python
n_epoch = 100
batch_size = 100
log_dir = "logs/trial1/"
tensorboard_callback = tf.keras.callbacks.TensorBoard(log_dir=log_dir, histogram_freq=1)
uvp_train = concatenate([u_train,v_train,p_train],1)
h = model.fit([t_train,x_train,y_train],[eta_train, uvp_train ], epochs=n_epoch, batch_size=batch_size, shuffle=True, callbacks = [tensorboard_callback])
```
